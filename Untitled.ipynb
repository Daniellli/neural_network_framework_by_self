```python
#返回node所连接的节点
def get_output(graph,node):
    outputs = []
    for n,links in graph.items():
        if node == n:
            outputs+=links
    return outputs


get_output(computing_graph,'k1')
```

### 如何获得K1的偏导

```python
def get_parameter_partial_order(target):
    computing_order=[]
    out = get_output(computing_graph,target)[0]
    computing_order.append(target)

    while out:
        computing_order.append(out)
        out = get_output(computing_graph,out)
        if out : out = out[0]
    order = []
    for index,n  in enumerate( computing_order[:-1]):
        order.append((computing_order[index+1],n))
    return '*'.join(["∂{}/∂{}".format(a,b) for a,b in order[::-1]])

```

```python
def topologic(graph):
    """
    test_graph = {
        x:[linear],
        k:[linear],
        b:[linear],
        
        linear:[sigmoid],
        sigmoid:[loss],
        y:[loss]
    }
    
    """
    sorted_node= []
    
    while graph:
        all_nodes_inputs  = reduce( lambda a,b: a+b,list(graph.values()))
        all_nodes_outputs = list(graph.keys())
        
        all_node_only_have_outputs_no_inputs = set(all_nodes_outputs) - set(all_nodes_inputs)
        
        if all_node_only_have_outputs_no_inputs:
            node = random.choice(list(all_node_only_have_outputs_no_inputs))
            sorted_node.append(node)
            if len(graph) ==1 :sorted_node +=graph[node]
            graph.pop(node)
            
            
            for _,links in graph.items():#好像不用把 ！！
                if node in links:links.remove(node)
        else :
            raise TypeError("this graph has circle,which can not get topologic order")
        
    return sorted_node
```

```python
class Node:
    def __init__(self,inputs=[],name = None,is_trainable=False ):
        self.inputs = inputs
        self.outputs = []
        self.name = name
        self.value = None
        self.gradients = dict()
        self.is_trainable = is_trainable
        
        for node in inputs:
            node.outputs.append(self)
    #__repr__修改输出结果
    def __repr__(self):
        return 'Node:{}'.format(self.name)
    
    #向前运算
    
    def forward(self):
        pass
#         print("i am {} , I have no human baba,I calculate  myself  value:{} by myself !!!".format(self.name,self.value))
        

    def backward(self):
        for n in self.inputs:
            print("get ∂{} / ∂{} ".format(self.name,n.name))
       
    
    
    
    
    
    
    
    
    
    
    
    
    
    
class Placeholder(Node):
    def __init__(self,name = None,is_trainable =False):
        Node.__init__(self,name = name,is_trainable=is_trainable )

    #__repr__修改输出结果
    def __repr__(self):
        return 'Node:{}'.format(self.name)
    
    #向前运算
    
    def forward(self):
        pass
#         print("i am {} , I was  assigned value:{} by human baba  !".format(self.name,self.value))
       
    
    def backward(self):
        self.gradients[self] = self.outputs[0].gradients[self]
#         print("i got myself gradients :{}".format(self.outputs[0].gradients[self]))
        

    
    
    
    
    
    
    
    
    
    
    
    
class Linear(Node):
    def __init__(self,x,k,b,name = None):
        Node.__init__(self,inputs=[x,k,b],name = name)
        
    #__repr__修改输出结果
    def __repr__(self):
        return 'Node:{}'.format(self.name)
    
    #向前运算
    
    def forward(self):
        
      
        x,k,b = self.inputs[0],self.inputs[1],self.inputs[2]
        self.value =  k.value *x.value  + b.value
#         print("i am {} , I have no human baba,I calculate  myself  value ：{}by myself !!!".format(self.name,self.value))

        
    def backward(self):
        #self.input[0]  也就是 x
        #self.input[1]  也就是 k
        #self.input[2]  也就是 b
        #self.outputs[0] means  sigmoid   
        #self.outputs[0].gradients[self]               means  sigmoid 对自己的导
        x,k,b = self.inputs[0],self.inputs[1],self.inputs[2]   
#         self.gradients[self.inputs[0]] =  '*'.join([self.outputs[0].gradients[self],  '∂{}/∂{}'.format(self.name,self.inputs[0].name)])
#         self.gradients[self.inputs[1]] =  '*'.join([self.outputs[0].gradients[self],  '∂{}/∂{}'.format(self.name,self.inputs[1].name)])
#         self.gradients[self.inputs[2]] =  '*'.join([self.outputs[0].gradients[self],  '∂{}/∂{}'.format(self.name,self.inputs[2].name)])
        self.gradients[self.inputs[0]] = self.outputs[0].gradients[self]* k.value
        self.gradients[self.inputs[1]] =  self.outputs[0].gradients[self] *x.value
        self.gradients[self.inputs[2]] = self.outputs[0].gradients[self] *1
#         print('self.gradient[self.input[0]] {} '.format(self.gradients[self.inputs[0]]) ) 
#         print('self.gradient[self.input[1]] {} '.format(self.gradients[self.inputs[1]]) ) 
#         print('self.gradient[self.input[2]] {} '.format(self.gradients[self.inputs[2]]) ) 
       
    
        
        
        
        
        
        
        
        
        
        
        
        
        
        
        
class Sigmoid(Node):
    def __init__(self,x,name = None):
        Node.__init__(self,inputs=[x],name = name)
        
    #__repr__修改输出结果
    def __repr__(self):
        return 'Node:{}'.format(self.name)
    
    
    def _sigmoid(self,x):
        return 1/(1+np.exp(-x))
        
    #向前运算
    
    def forward(self):
        
       
        x = self.inputs[0]
        self.value =  self._sigmoid(x.value)
#         print("i am {} , I have no human baba,I calculate  myself  value ：{}by myself !!!".format(self.name,self.value))
        
    def backward(self):
        #self.input[0]  means  linear 
        #self.gradients[self.input[0]]       means      sigmoid对 linear的导
        #self.outputs[0]     means        loss  
        #join(self.outputs[0].gradients[self]      means     loss对 自己 也就是sigmoid的导  
            
#         self.gradients[self.inputs[0]] =  '*'.join([self.outputs[0].gradients[self],  '∂{}/∂{}'.format(self.name,self.inputs[0].name)])
       #∂(x)的偏导 ==  ∂(x) * (1 - ∂(x))
        x = self.inputs[0].value
        self.gradients[self.inputs[0]] =  self.outputs[0].gradients[self]  *  self._sigmoid(x)*(1 - self._sigmoid(x))
#         print('self.gradient[self.input[0]] {} '.format(self.gradients[self.inputs[0]]) ) 
       
    
        
        
        
        
        
        
        
        
        
class Loss(Node):
    def __init__(self,y,yhat,name= None):
        Node.__init__(self,inputs=[y,yhat],name = name)
        
    #__repr__修改输出结果
    def __repr__(self):
        return 'Node:{}'.format(self.name)
    
    #向前运算
    
    def forward(self):
        
        y,yhat = self.inputs[0],self.inputs[1]
        self.value =  np.mean((y.value-yhat.value)**2)
#         print("i am {} , I have no human baba,I calculate  myself  value ：{}by myself !!!".format(self.name,self.value))
     
    def backward(self):
        ##self.input[0] 就是y ,
        #self.input[1] 就是yhat
        #self.gradients[self.input[0]] means loss 对 y的偏导
        # self.gradients[self.input[1]]  means loss 对 yhat的偏导
        
#         self.gradients[self.inputs[0]] = '∂{}/∂{}'.format(self.name,self.inputs[0].name)
#         self.gradients[self.inputs[1]] = '∂{}/∂{}'.format(self.name,self.inputs[1].name)
        y,yhat = self.inputs[0],self.inputs[1]
        self.gradients[self.inputs[0]] = 2 *np.mean(y.value - yhat.value)
        self.gradients[self.inputs[1]] = -2 *np.mean(y.value - yhat.value)
        
        
#         print('self.gradient[self.input[0]] {} '.format(self.gradients[self.inputs[0]]) ) 
#         print('self.gradient[self.input[1]] {} '.format(self.gradients[self.inputs[1]]))
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
#Our simple model elements

#version 1
# node_x = Node(input=None,output = [node_linear])
# node_y = Node(input=None,output = [node_loss])
# node_k = Node(input=None,output = [node_linear])
# node_b = Node(input=None,output = [node_linear]) 
# node_linear = Node(input=[node_x,node_k,node_b],output = [node_sigmoid])
# node_sigmoid = Node(input=[node_linear],output = [node_loss])
# node_loss = Node(input=[node_sigmoid,node_y],output = None)
#version 2
node_x = Placeholder(name='x')
node_y = Placeholder(name='y')
node_k = Placeholder(name='k',is_trainable = True)
node_b = Placeholder(name = 'b',is_trainable = True) 
node_linear = Linear(node_x,node_k,node_b,name='linear')
node_sigmoid = Sigmoid(x = node_linear,name='sidmoid')
node_loss = Loss(yhat = node_sigmoid,y = node_y,name = 'loss')
```

```python
#将feed_dict  只有边界节点  需要扩充成完整图

def convert_feed_dict_graph(feed_dict):
    need_expand  = [n for n in feed_dict]
    computing_graph = defaultdict(list)  #存储最终图

    #当还有边界节点没扩充
    while need_expand:
        node_pop = need_expand.pop(0) #随机弹一个边界节点

        if node_pop in computing_graph:  #已经扩充完了  不用扩充了 
            continue
        #给边界点赋值
        if isinstance(node_pop,Placeholder):node_pop.value = feed_dict[node_pop]
            
        
        #当前弹出的边界节点 的所以出边 都加入边界节点的结合中  
        #并且加到当前弹出节点的出边表中
        for m in node_pop.outputs: 
            computing_graph[node_pop].append(m)
            need_expand.append(m)
            
    return computing_graph

feed_dict = {
    node_x:3,
    node_y:random.random(),
    node_k:random.random(),
    node_b:0.38,
    
} 

sorted_nodes = topologic(convert_feed_dict_graph(feed_dict)  )
sorted_nodes
```

##  经过拓扑排序就一定能够实现 后续节点在后面计算

```python
#前向传播
def forward_propagation(nodes):
    for node in nodes:
        node.forward()
#         if isinstance(node,Loss):
#             print("loss value  = {}".format(node.value))
    
def backwar_propagation(nodes):
    for node in sorted_nodes[::-1]:
#         print("\n i am:{}".format(node.name))
        node.backward()
    

def run_one_epoch(graph):
    forward_propagation(graph)
    backwar_propagation(graph)
    
def optimizer(graph,learning_rate = 1e-1):
    for node in graph:
        if node.is_trainable:
            node.value = node.value + -1*node.gradients[node] *learning_rate
#             cmp = 'large' if node.gradients[node] >0 else 'small'
#             print("{} value is too {},I need update myself".format(node.name,cmp))

```

```python
def _core(sorted_nodes,iterations,learning_rate=1e-1):
    loss_history = []
    for i in range(iterations):
        run_one_epoch(sorted_nodes)
        _loss = sorted_nodes[-1]
        assert isinstance(_loss,Loss)
        loss_history.append(_loss.value)

        optimizer(sorted_nodes)
    return sorted_nodes,loss_history
    # loss_history
```
