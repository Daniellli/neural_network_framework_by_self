 # Mean Squared Error :MSE  平方误差函数

```python

import random
```

```python
def model(x,k,b):
    return x*k+b
```

```python
def loss(y_hat, y):
    return np.mean((y_hat - y) ** 2)
```

```python

K_MAX,VAR_MIN = 100,-100

min_loss = float("inf")#无穷大

best_k,best_b = None,None
total_times = 1000


for t in range(total_times):
    k,b = random.randint(VAR_MIN,K_MAX),random.randint(VAR_MIN,K_MAX)
    loss_  = loss(Y,model(X_rm,k,b))
    if loss_ <min_loss:
        min_loss = loss_
        best_k ,best_b = k,b
        print("在{}时刻我找到了更好的K：{}和b:{}，这个时候的loss是 :{}".format(t,k,b,loss_))
```

```python
def partial_k(x,y,k,b):
    return 2*np.mean((y-(k*x+b))*(-x))


def partial_b(x,y,k,b):
    return 2*np.mean((y-(k*x+b))*(-1))


```

# 深度学习的核心 ：通过梯度下降的方法，获取一组参数，使得loss最小

```python

# K_MAX,VAR_MIN = 100,-100

min_loss = float("inf")

best_k,best_b = None,None
total_times = 1000
alpha = 1e-3

for t in range(total_times):
    
    c_k=  c_k+(-1)*partial_k(X_rm,Y,c_k,c_b ) * alpha
    c_b = c_b+(-1)*partial_b(X_rm,Y,c_k,c_b )* alpha
#     k,b = random.randint(VAR_MIN,K_MAX),random.randint(VAR_MIN,K_MAX)
    loss_  = loss(Y,model(X_rm,c_k,c_b))
    if loss_ <min_loss:
        min_loss = loss_
        best_k ,best_b = c_k,c_b
        print("在{}时刻我找到了更好的K：{}和b:{}，这个时候的loss是 :{}".format(t,c_k,c_b,loss_))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
def sigmoid(x):
    return 1/ (1+ np.exp(-x))
```

```python
sub_x = np.linspace(-10,10)
plt.plot(sub_x,sigmoid(sub_x))
```

```python
def random_linear(x):
    k,b = random.random(),random.random()
    return k*x +b

```

```python

sub_x = np.linspace(-10,10)
for _ in range(5):
    plt.plot(sub_x,random_linear(sigmoid(random_linear(sub_x))))
```
